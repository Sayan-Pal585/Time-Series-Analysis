```python
# import packages
import pandas as pd
import warnings
import plotly
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
warnings.simplefilter("ignore")
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import plotly.graph_objs as go
from plotly.offline import plot,init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
# reading csv file into dataframe
# Set path of input file
df = pd.read_csv("C:\\Users\\SurabhiGupta\\Desktop\\Daimler_Scripts\\Scripts\\Forecasting Scripts\\forecast\\forecast\\Input Files\\Al Prices.csv")
print(df.head())

# Convert Date column to date format
df['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')

# Setting Date column as index
df.set_index('Date', inplace=True)
print(df.tail())
```

```python
# Recommended method 3
# ARIMA (1,2,1)

accuracy = pd.read_csv("C:\\Users\\SurabhiGupta\\Desktop\\Daimler_Scripts\\Scripts\\Forecasting Scripts\\forecast\\forecast\\Input Files\\Accuracy_Input.csv")


# Setting date to index
accuracy['Snap_Month'] = pd.to_datetime(accuracy['Snap_Month'],format='%d-%m-%Y')
accuracy.set_index('Snap_Month', inplace=True)
print(accuracy)


from mathmodels.forecast import arima

for i in range(108,133):
    dfTrain=df[:i]
    dfTest=df[i:]
    print(dfTrain)
    print(dfTest)
    
    # Building and fitting model for 2 year test set
    arimaModel = dfTrain.arima.build('Prices', order = (1,2,1))

    # Model Summary
    arimaSummary = dfTrain.arima.getSummary(arimaModel)
    print(arimaSummary)

    # scoring model on training set
    arimaDfTrain = dfTrain.arima.scoreTrain(arimaModel, 'Prices')
    print(arimaDfTrain)

    # forecast test set values
    forecast1 = arimaModel.forecast(24)
    print(forecast1)
    forecast1 = pd.DataFrame({'Prediction': forecast1[0]})
    forecast1.set_index(accuracy.index[i-107:i-107+24],inplace=True)
    print(forecast1)
    accuracy=pd.merge(accuracy,forecast1,how='outer',left_index=True,right_index=True)
    print(accuracy)

accuracy.columns = ['A','F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12','F13','F14','F15','F16','F17','F18','F19','F20','F21','F22','F23','F24','F25']
print(accuracy)    
accuracy.to_csv("C:\\Users\\SurabhiGupta\\Desktop\\Daimler_Scripts\\Scripts\\Forecasting Scripts\\forecast\\forecast\\Output Files\\Accuracy_Input_Al_3_ARIMA_121.csv")    

```

```python
# Recommended Method- 2
# ARIMA for USDI_lag1 and ARIMAX with USDI_lag1 & Event

from mathmodels.forecast import arima

accuracy = pd.read_csv("C:\\Users\\SurabhiGupta\\Desktop\\Daimler_Scripts\\Scripts\\Forecasting Scripts\\forecast\\forecast\\Input Files\\Accuracy_Input.csv")


# Setting date to index
accuracy['Snap_Month'] = pd.to_datetime(accuracy['Snap_Month'],format='%d-%m-%Y')
accuracy.set_index('Snap_Month', inplace=True)
print(accuracy)

   
# reading exogenous file into dataframe
# Set path of input file
df_X = pd.read_csv("C:\\Users\\SurabhiGupta\\Desktop\\Daimler_Scripts\\Scripts\\Forecasting Scripts\\forecast\\forecast\\Input Files\\exogenous.csv")
print(df_X.head())

# Convert Date column to date format
df_X['Date'] = pd.to_datetime(df_X['Date'],format='%d-%m-%Y')

# Setting Date column as index
df_X.set_index('Date', inplace=True)
print(df_X.head())

for i in range(108,133):
    dfTrain=df[:i]
    dfTest=df[i:]
    print(dfTrain)
    print(dfTest)
    
    # Split Exog into Training & Test Datasets
    dfTrain_X = df_X[:i]
    dfTest_X = df_X[i:]
    print(dfTrain_X)
    print(dfTest_X)


    # Split Exog into Training & Test Datasets for ARIMA modelling for exog 
    exogTrain=dfTrain_X[['USDI_lag1','Event']]
    print(exogTrain)
    exogTest=dfTest_X[['USDI_lag1','Event']]
    print(exogTest)

    # ARIMA for Exogenous
    exogarimaModel = exogTrain.arima.build('USDI_lag1', order = (1,2,1))

    # Exog Model Summary
    exogarimaSummary = exogTrain.arima.getSummary(exogarimaModel)
    print(exogarimaSummary)
    
    # exog forecasted 24 values
    forecast1 = exogarimaModel.forecast(24)
    print(forecast1)
    forecast1 = pd.DataFrame({'predicted': forecast1[0]})
    print(forecast1)
    forecast1.set_index(accuracy.index[i-107:i-107+24],inplace=True)
    print(forecast1)
    
    exogarimaforecast=pd.merge(forecast1,exogTest,how='left',left_index=True,right_index=True)
    print('exogarimaforecast')
    print(exogarimaforecast)
    exogarimaforecast['Event']=pd.np.where(exogarimaforecast['Event'].isna(),0,exogarimaforecast['Event'])
    print(exogarimaforecast)
    
    
    # ARIMAX for Al prices
    # Building and fitting model 
    arimaModel = dfTrain.arima.build('Prices', order = (1,0,1),exog=exogTrain[['USDI_lag1','Event']])

    # Model Summary
    arimaSummary = dfTrain.arima.getSummary(arimaModel)
    print(arimaSummary)

    # scoring model on training set
    arimaDfTrain = dfTrain.arima.scoreTrain(arimaModel, 'Prices')
    print(arimaDfTrain)

    # Forecasted values for Prices 
    Forecast = arimaModel.forecast(24,exog=exogarimaforecast[['predicted','Event']])
    Forecast = pd.DataFrame({'Prediction': Forecast[0]})
    # Setting index
    Forecast.set_index(accuracy.index[i-107:i-107+24],inplace=True)
    print(Forecast)
    accuracy=pd.merge(accuracy,Forecast,how='outer',left_index=True,right_index=True)
    print(accuracy)

accuracy.columns = ['A','F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12','F13','F14','F15','F16','F17','F18','F19','F20','F21','F22','F23','F24','F25']
print(accuracy)    
accuracy.to_csv("C:\\Users\\SurabhiGupta\\Desktop\\Daimler_Scripts\\Scripts\\Forecasting Scripts\\forecast\\forecast\\Output Files\\Accuracy_Input_Al_2_ARIMAX_USDI_lag1_Event.csv")    

  




```
