```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from requests import get
from datetime import datetime as dt
```

### Extracting latest news articles for each commodity

```python
df=pd.DataFrame(columns=['Metal','Title','Headline','Ref Link','Location','Date','Time','Date-Time'])



def news(metal,url,id1,df):
    
    response = get(url)

    html_soup = BeautifulSoup(response.text)

    for i in html_soup.find_all('div',id=id1):
        for j in i.find_all('ul'):
            for k in j.find_all('li'):  
                location='Not Available'
                datetime='Not Available'
                date_only = 'Not Available'
                time_only = 'Not Available'
                for l in k.find_all('div',class_="headline-title"):
                    for m in l.find_all('a'):
                        title=m.get_text()
                        title_headline = title.split(':')
                        title = title_headline[0]
                        if(len(title_headline) == 1):
                            headline =  title_headline[0]
                        elif(len(title_headline)==2):
                            headline = title_headline[1]
                        else:
                            for th in range(1,len(title_headline)):
                                headline += title_headline[th]
                        ##print('title',title)
                        link=m['href']
                        ##print('link',link)
                        link1='https://metalbulletin.com'+link
                        ##print('link1',link1)                
                for p in k.find_all('span',class_="headline-time"):
                    location=p.get_text()
                    ##print('location',location)
                for q in k.find_all('span',class_="headline-location"):
                    datetime=q.get_text()
                    dtime = datetime.split('@')
                    
                    if(len(dtime)>=2):
                        date_only = dtime[0].strip(" ")
                        date_only = date_only.strip("\n")
                        date_only = dt.strptime(date_only, '%b %d, %Y').strftime('%d-%m-%Y')
                        time_only = dtime[1].strip(" ")
                        time_only = time_only.strip("\n")
                        datetime = date_only+' '+dtime[1]
                    else:
                        datetime = 'Not Available'

                df=df.append({'Metal':metal,'Title':title,'Headline':headline,'Ref Link':link1,'Date':date_only,'Time':time_only,'Date-Time':datetime,'Location':location},ignore_index=True)
                


    df = df.replace('\\n','', regex=True)
    df = df.replace('\t','', regex=True)

    return df



df=news('Aluminium','https://www.metalbulletin.com/non-ferrous/base-metals/aluminium.html','2665400',df)
df=news('Nickel','https://www.metalbulletin.com/non-ferrous/base-metals/nickel.html','2665504',df)
df=news('Cobalt','https://www.metalbulletin.com/non-ferrous/minor-metals/cobalt.html','2665935',df)
df=news('Steel','https://www.metalbulletin.com/steel.html','2668623',df)





#Lithium
metal='Lithium'
url='https://www.metalbulletin.com/non-ferrous/battery-raw-materials/lithium.html'
response = get(url)
html_soup = BeautifulSoup(response.text)

for i in html_soup.find_all('section',id="metal-news"):
    for j in i.find_all('ul'):
        for jk in j.find_all('li',limit=1):
            for k in jk.find_all('li'):
                location='Not Available'
                datetime='Not Available'
                date_only = 'Not Available'
                time_only = 'Not Available'
                for l in k.find_all('div',class_="headline-title"):
                    for m in l.find_all('a'):
                        title=m.get_text()
                        title_headline = title.split(':')
                        title = title_headline[0]
                        if(len(title_headline) == 1):
                            headline =  title_headline[0]
                        elif(len(title_headline)==2):
                            headline = title_headline[1]
                        else:
                            for th in range(1,len(title_headline)):
                                headline += title_headline[th]
                        #print('title',title)
                        link=m['href']
                        #print('link',link)
                        link1='https://metalbulletin.com'+link
                        #print('link1',link1)                
                for p in k.find_all('span',class_="headline-time"):
                    location=p.get_text()
                    #print('location',location)
                for q in k.find_all('span',class_="headline-location"):
                    datetime=q.get_text()
                    dtime = datetime.split('@')
                    
                    if(len(dtime)>=2):
                        date_only = dtime[0].strip(" ")
                        date_only = date_only.strip("\n")
                        date_only = dt.strptime(date_only, '%b %d, %Y').strftime('%d-%m-%Y')
                        time_only = dtime[1].strip(" ")
                        time_only = time_only.strip("\n")
                        datetime = date_only+' '+time_only
                    else:
                        datetime = 'Not Available'
                    #print('time',datetime)

                
                df=df.append({'Metal':metal,'Title':title,'Headline':headline,'Ref Link':link1,'Date':date_only,'Time':time_only,'Date-Time':datetime,'Location':location},ignore_index=True)
            
                


df = df.replace('\\n','', regex=True)
df = df.replace('\t','', regex=True)




#Rubber
metal='Rubber'
url='https://globalrubbermarkets.com/global-rubber-market-news'
response = get(url)
html_soup = BeautifulSoup(response.text)

for i in html_soup.find_all('div',class_='td_block_inner'):
    for j in i.find_all('div',class_='td-big-grid-wrapper'):
        for k in j.find_all('div'):
            if(k['class'] == ['td-big-grid-scroll'] or k['class'] == ['td_module_mx5', 'td-animation-stack', 'td-big-grid-post-0', 'td-big-grid-post', 'td-big-thumb']):
                location='Not Available'
                datetime='Not Available'
                for l in k.find_all('div',class_="td-module-thumb"):
                    for m in l.find_all('a'):
                        #print('Rubber')
                        title=m.get('title')
                        title_headline = title.split(':')
                        title = title_headline[0]
                        if(len(title_headline) == 1):
                            headline =  title_headline[0]
                        elif(len(title_headline)==2):
                            headline = title_headline[1]
                        else:
                            for th in range(1,len(title_headline)):
                                headline += title_headline[th]
                        #print('title',title)
                        link=m.get('href')
                        #print('link',link)
                        url=link
                        response = get(url)
                        html_soup = BeautifulSoup(response.text)
                        for b in html_soup.find_all('div',class_='td-post-header'):
                            for c in b.find_all('header',class_='td-post-title'):
                                for d in c.find_all('span',class_='td-post-date'):
                                    for e in d.find_all('time'):
                                        date_time=e.get('datetime')
                                        #date_only = dt.strptime(date_time, '%Y-%m-%dT%H:%M:%S+00:00').strftime('%b %d, %Y')
                                        date_only = dt.strptime(date_time, '%Y-%m-%dT%H:%M:%S+00:00').strftime('%d-%m-%Y')
                                        date_only = str(date_only)
                                        time_only = dt.strptime(date_time, '%Y-%m-%dT%H:%M:%S+00:00').strftime('%H:%M')
                                        time_only = str(time_only)
                                        #date_time_only = dt.strptime(date_time, '%Y-%m-%dT%H:%M:%S+00:00').strftime('%b %d, %Y %H:%M')
                                        date_time_only = date_only+" "+time_only
                                        #print(location)
                        
      
                        df=df.append({'Metal':metal,'Title':title,'Headline':headline,'Ref Link':link,'Date':date_only,'Time':time_only,'Date-Time':date_time_only,'Location':'Not Available'},ignore_index=True)
                        
                


df = df.replace('\\n','', regex=True)
df = df.replace('\t','', regex=True)
#print(df)

                        
#df.to_csv('Metal_news_12.csv',index=False)
```

### Converting to one date time format

```python
df['Date'] = pd.to_datetime(df['Date'],infer_datetime_format=True)
df['Date-Time'] = pd.to_datetime(df['Date-Time'],infer_datetime_format=True)
```

### Opening previous articles which is stored in News_collated_final.csv sheet

```python
df_total = pd.read_csv('News_collated_final_sheet_16012020.csv')
df_total.head()
```

```python
df_total['Date'] = pd.to_datetime(df_total['Date'],infer_datetime_format=True)
df_total['Date-Time'] = pd.to_datetime(df_total['Date-Time'],infer_datetime_format=True)
```

```python
df_total.drop(['index'],axis=1,inplace=True)
```

### Concating previous and today's articles and removing articles which already exist

```python
df_total = pd.concat([df,df_total],ignore_index=True)
df_total.drop_duplicates(['Metal','Title','Ref Link'],keep='last',inplace=True)
df_total.reset_index().drop(['index'],axis=1,inplace=True)
df_total.head()
```

### Ranking news articles based on time

```python
Metals = ['Aluminium','Steel','Nickel','Cobalt','Lithium','Rubber']
key = 1
for i in Metals:
    df_test = df_total[df_total['Metal'] == i].reset_index().copy(deep=True)
    df_test.drop(columns=['index'],axis=1,inplace=True)
    if(key):
        df_grouped = df_test.copy(deep=True)
        key = 0
    else:
        df_grouped = pd.concat([df_grouped,df_test],ignore_index=False)
df_grouped.reset_index(inplace=True)
```

### saving in file
